Nama kolom untuk dataset didefinisikan dalam daftar columns, yang mencakup lima kolom: sepal_length, sepal_width, petal_length, petal_width, dan class, yang mewakili spesies bunga iris (Setosa, Versicolor, atau Virginica). Fungsi pd.read_csv digunakan untuk mengunduh data dari URL tersebut dan memuatnya ke dalam DataFrame df, dengan menetapkan parameter header=None untuk menunjukkan bahwa dataset tidak memiliki header baris, dan names=columns untuk memberikan nama kolom sesuai dengan yang telah didefinisikan. 1. Clustering menggunakan dataset iris menggunakan fitur 2 dan 3

```python
import matplotlib.pyplot as plt

f1 = df['sepal_width'].values
f2 = df['petal_length'].values
x = np.array(list(zip(f1, f2)))
plt.scatter(f1, f2, c='black', s=50)
```

Kode di atas digunakan untuk memvisualisasikan data dalam bentuk scatter plot dengan menggunakan dua fitur dari dataset Iris, yaitu *sepal width* (lebar kelopak) dan *petal length* (panjang kelopak). Pertama, nilai dari kolom `sepal_width` dan `petal_length` diambil dan disimpan dalam variabel `f1` dan `f2` masing-masing. Kedua variabel ini kemudian dipasangkan menjadi pasangan koordinat (x, y) dengan menggunakan fungsi `zip()`, yang menggabungkan nilai dari kedua fitur tersebut. Hasilnya, sebuah array numpy `x` terbentuk, yang menyimpan pasangan koordinat dari setiap titik data. Selanjutnya, fungsi `plt.scatter()` digunakan untuk menggambar scatter plot, di mana nilai `f1` (sepal width) diplot pada sumbu x, dan nilai `f2` (petal length) diplot pada sumbu y. Titik-titik pada plot diberi warna hitam (`c='black'`), dengan ukuran titik yang cukup besar, yaitu 50 (`s=50`), agar lebih terlihat jelas. Visualisasi ini memberikan gambaran tentang bagaimana dua fitur tersebut tersebar dalam dataset Iris dan apakah terdapat pola atau hubungan yang jelas di antara keduanya.

```python
plt.title("Plotting Data Unsupervised")
plt.scatter(f1, f2, c='black', s=50)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.grid()
plt.show()
```

Kode di atas bertujuan untuk memperkaya visualisasi scatter plot yang menggambarkan hubungan antara dua fitur dari dataset Iris, yaitu sepal_width dan petal_length, dengan menambahkan beberapa elemen penting untuk meningkatkan pemahaman terhadap grafik. Pertama, fungsi plt.title("Plotting Data Unsupervised") digunakan untuk memberi judul pada plot, yang mengindikasikan bahwa grafik ini digunakan untuk eksplorasi data tanpa supervisi, atau unsupervised learning. Selanjutnya, plt.scatter(f1, f2, c='black', s=50) menghasilkan scatter plot dari data, dengan titik-titik yang diwarnai hitam dan ukuran titik sebesar 50 piksel. Agar lebih informatif, sumbu x dan y diberi label menggunakan plt.xlabel("sepal_width") dan plt.ylabel("petal_length"), yang menjelaskan variabel apa yang masing-masing sumbu representasikan. Untuk meningkatkan keterbacaan, fungsi plt.grid() ditambahkan untuk menampilkan grid pada plot, memudahkan pembaca untuk membaca nilai-nilai pada sumbu x dan y.
```python
from sklearn.cluster import KMeans

# Initializing KMeans
kmeans = KMeans(n_clusters=3)
# Fitting with inputs
kmeans = kmeans.fit(x)
# predicting the clusters
labels = kmeans.predict(x)
# Getting the clusters centers
C = kmeans.cluster_centers_

print(C)
```

Kode di atas menggunakan algoritma *KMeans* dari pustaka `scikit-learn` untuk melakukan klastering data Iris dengan tujuan membagi data ke dalam 3 klaster berdasarkan dua fitur, yaitu *sepal_width* dan *petal_length*. Pertama, objek `KMeans` diinisialisasi dengan parameter `n_clusters=3`, yang menentukan jumlah klaster yang ingin dibentuk. Selanjutnya, model *KMeans* dilatih menggunakan metode `fit(x)`, di mana `x` adalah data yang berisi dua fitur tersebut. Setelah model terlatih, fungsi `predict(x)` digunakan untuk memprediksi klaster yang sesuai untuk setiap titik data, dan hasil prediksi tersebut disimpan dalam variabel `labels`. Terakhir, koordinat titik pusat (centroid) dari setiap klaster dihitung dengan menggunakan atribut `cluster_centers_`, yang menghasilkan posisi rata-rata dari data yang ada di dalam klaster tersebut. Hasilnya, kita mendapatkan koordinat titik pusat untuk masing-masing klaster, yang kemudian dicetak dengan `print(C)`. Titik pusat ini menggambarkan posisi rata-rata data dalam klaster dan memberikan informasi tentang "pusat" dari masing-masing klaster dalam ruang fitur yang digunakan.

```python
plt.scatter(x[:,0], x[:,1], s=50, c=labels)
plt.scatter(C[:,0], C[:,1], marker='*', s=200, c='#050505')
```

Kode ini digunakan untuk memvisualisasikan hasil klastering data menggunakan algoritma *KMeans*, di mana data yang telah dikelompokkan akan dipetakan dalam bentuk scatter plot, dan titik pusat (centroid) dari setiap klaster juga akan ditampilkan. Pada bagian pertama kode, `plt.scatter(x[:,0], x[:,1], s=50, c=labels)` digunakan untuk menggambar scatter plot dari data yang telah dikelompokkan. Data yang dipisahkan menjadi dua fitur, misalnya *sepal width* dan *petal length*, dipetakan pada sumbu x dan y. Titik-titik data diberi warna yang berbeda sesuai dengan label klaster yang dihasilkan oleh *KMeans* melalui argumen `c=labels`. Hal ini memungkinkan kita untuk melihat bagaimana data tersebar dan dikelompokkan berdasarkan klaster yang berbeda. Pada bagian kedua kode, `plt.scatter(C[:,0], C[:,1], marker='*', s=200, c='#050505')`, digunakan untuk memplot titik pusat (centroid) dari setiap klaster. Titik-titik pusat ini ditandai dengan simbol bintang (`marker='*'`) dan diberi warna merah tua, dengan ukuran yang lebih besar agar lebih mencolok. Titik pusat ini menggambarkan posisi rata-rata dari setiap klaster dan memberikan informasi visual yang jelas tentang lokasi pusat klaster di dalam ruang data. Secara keseluruhan, visualisasi ini membantu untuk memahami bagaimana data dikelompokkan dalam klaster-klaster yang berbeda dan menunjukkan posisi rata-rata (centroid) dari setiap klaster dalam plot.

2. Evaluasi hasil clustering menggunakan inertia silhoutte coefficient

```python
SSE = []
for k in range(1, 10):
    # Define the number of clusters
    kmeans = KMeans(n_clusters=k, random_state=1)
    # Fit the KMeans model with the input data
    kmeans.fit(x)
    # Obtain cluster labels
    labels = kmeans.predict(x)
    # Calculate the sum of squared distances (SSE) for inertia
    inertia = kmeans.inertia_
    SSE.append(inertia)
    # Print the number of clusters and the inertia
    print(f"n-cluster = {k} , cost = {inertia}")
```

Kode di atas digunakan untuk menghitung dan menganalisis *Sum of Squared Errors* (SSE) atau *Inertia* pada klastering menggunakan algoritma *KMeans* untuk berbagai jumlah klaster yang berbeda. Proses dimulai dengan mendefinisikan sebuah daftar kosong, `SSE`, untuk menyimpan nilai *inertia* yang dihitung pada setiap iterasi. Kemudian, sebuah loop `for` digunakan untuk menguji berbagai jumlah klaster mulai dari 1 hingga 9. Pada setiap iterasi, model *KMeans* diinisialisasi dengan jumlah klaster yang berbeda (ditentukan oleh variabel `k`) dan dilatih menggunakan data input `x` dengan fungsi `fit(x)`. Setelah model dilatih, fungsi `predict(x)` digunakan untuk memprediksi label klaster untuk setiap data, sementara `kmeans.inertia_` menghitung nilai *inertia*, yang merupakan jumlah jarak kuadrat dari setiap titik data ke pusat klaster terdekat. Nilai *inertia* ini kemudian ditambahkan ke dalam daftar `SSE`, dan hasilnya dicetak untuk memberikan informasi mengenai jumlah klaster yang digunakan dan nilai biaya yang terkait dengan klaster tersebut. Proses ini memungkinkan kita untuk melihat bagaimana nilai *inertia* berubah saat jumlah klaster meningkat, dan ini dapat membantu dalam memilih jumlah klaster optimal, terutama ketika diterapkan dengan metode *elbow* untuk menentukan titik di mana penurunan *inertia* mulai melambat.

3. Jumlah K dengan elbow method

```python
k = range(1,10)
plt.plot(k, SSE, 'bx-')
plt.xlabel('Jumlah Cluster')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

Kode di atas digunakan untuk menggambarkan hasil perhitungan *Sum of Squared Errors* (SSE) atau *Inertia* yang dihitung sebelumnya dalam bentuk grafik menggunakan metode *elbow method*. Grafik ini bertujuan untuk membantu menentukan jumlah klaster optimal untuk algoritma *KMeans* berdasarkan perubahan nilai SSE seiring dengan meningkatnya jumlah klaster.

Langkah pertama, `k = range(1,10)`, mendefinisikan rentang jumlah klaster yang akan diuji, yaitu dari 1 hingga 9. Rentang ini digunakan untuk sumbu x dalam grafik, yang mewakili jumlah klaster yang diuji. Selanjutnya, `plt.plot(k, SSE, 'bx-')` menggambar grafik garis dengan titik data berbentuk silang (ditandai dengan `'bx-'`), di mana sumbu x adalah jumlah klaster (`k`), dan sumbu y adalah nilai SSE yang dihitung sebelumnya yang disimpan dalam daftar `SSE`. Dengan menggunakan garis dan titik, grafik ini akan memperlihatkan bagaimana SSE berubah saat jumlah klaster bertambah.

Kemudian, `plt.xlabel('Jumlah Cluster')` dan `plt.ylabel('SSE')` memberi label pada sumbu x dan y, masing-masing dengan "Jumlah Cluster" dan "SSE", untuk menjelaskan apa yang diwakili oleh kedua sumbu tersebut. `plt.title('The Elbow Method showing the optimal k')` memberikan judul pada grafik, yang menunjukkan bahwa grafik ini menggunakan metode siku (elbow method) untuk mencari jumlah klaster yang optimal. Titik siku pada grafik menunjukkan jumlah klaster di mana penurunan SSE mulai melambat secara signifikan, yang sering kali dianggap sebagai jumlah klaster yang terbaik.

Terakhir, `plt.show()` digunakan untuk menampilkan grafik yang telah dibuat. Dengan grafik ini, kita bisa dengan mudah mengidentifikasi jumlah klaster optimal berdasarkan titik siku, di mana penurunan SSE mulai melambat, menandakan bahwa menambah klaster lebih banyak tidak akan memberikan peningkatan signifikan dalam kualitas klastering.

4. Plot hasil cluster dengan nilai K

```python
# Apply KMeans with the chosen K (e.g., K=3)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(x)
labels = kmeans.labels_

# Scatter plot for the clusters
plt.figure(figsize=(8, 6))
for cluster in range(3):  # Adjust range based on the chosen K
    plt.scatter(x[labels == cluster, 0], x[labels == cluster, 1], label=f'Cluster {cluster + 1}')

# Plot centroids
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], marker='*', color='red', s=200, label='Centroids')

plt.title("Clusters of Iris Data")
plt.xlabel("Sepal Width")
plt.ylabel("Petal Length")
plt.legend()
plt.grid()
plt.show()
```

Kode di atas digunakan untuk menerapkan algoritma *KMeans* dengan jumlah klaster yang telah dipilih, yaitu 3, pada dataset Iris dan memvisualisasikan hasil klastering dalam bentuk scatter plot. Proses dimulai dengan menginisialisasi model *KMeans* dengan parameter `n_clusters=3` dan melatih model menggunakan data `x`. Setelah model terlatih, label klaster yang dihasilkan oleh *KMeans* disimpan dalam variabel `labels`. Untuk visualisasi, sebuah grafik scatter dibuat, di mana setiap klaster digambarkan dengan warna yang berbeda. Dalam hal ini, sebuah loop digunakan untuk memplot data berdasarkan label klaster yang dihasilkan, sehingga titik-titik data yang termasuk dalam klaster yang sama akan berada pada posisi yang sama di grafik, dengan warna yang konsisten. Selain itu, titik pusat (centroid) dari setiap klaster juga ditampilkan dengan simbol bintang merah yang lebih besar, untuk memudahkan identifikasi posisi rata-rata dari setiap klaster. Grafik ini diberi judul "Clusters of Iris Data", dengan label sumbu x "Sepal Width" dan sumbu y "Petal Length". Fungsi legenda digunakan untuk memberi penjelasan mengenai warna yang mewakili setiap klaster dan untuk menandakan titik pusat sebagai 'Centroids'. Grafik ini akhirnya dilengkapi dengan grid untuk memudahkan pembacaan, dan ditampilkan dengan fungsi `plt.show()`. Hasil visualisasi ini memberikan gambaran yang jelas mengenai distribusi data Iris dalam tiga klaster yang berbeda serta posisi pusat klaster tersebut.
